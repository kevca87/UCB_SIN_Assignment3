## Analysis about the expanded states when the algorithm prune and doesn't prune the states

```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import ndarray
```

```python
df_no_prunning = pd.read_csv('OthelloAdv_v1_loop_PC1_PC2_count_tiles_NoPrunning_d4.csv')
```

```python
df_no_prunning
```

```python
df_prunning = pd.read_csv('OthelloAdv_v1_loop_PC1_PC2_count_tiles_d4.csv')
```

```python
df_prunning
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('Game instant')  # Add an x-label to the axes.
ax.set_ylabel('States expanded')  # Add a y-label to the axes.
ax.set_title(f"Quantity of states expanded (Prunning vs. No prunning)")  # Add a title to the axes.
ax.set(xticks=[])
ax.plot(df_prunning['Movements'],df_no_prunning['Actions expanded'], label='No prunning')
ax.plot(df_prunning['Movements'],df_prunning['Actions expanded'], label='Prunning')
plt.show()
```
